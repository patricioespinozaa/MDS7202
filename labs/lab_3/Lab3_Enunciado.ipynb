````python
def to_snake_case(s: str) -> str:
    """Transforma un string a snake_case:
    los espacios se reemplazan por guiones bajos y todas las letras se convierten a minúsculas."""
    return s.lower().replace(" ", "_")
````

````python
def clean_data(df: pd.DataFrame) -> pd.DataFrame:
    # 1. Renombrar las columnas a snake_case
    df.columns = [to_snake_case(col) for col in df.columns]

    # 2. Eliminar race/ de la columna race/ethnicity
    df.rename(columns={"race/ethnicity": "ethnicity"}, inplace=True)

    # 3. Eliminar duplicados
    df.drop_duplicates(inplace=True)

    # 4. Pasar a float las columnas de notas con coerce
    for col in ["math_score", "reading_score", "writing_score"]:
        df[col] = pd.to_numeric(df[col], errors="coerce")

    # 5. Eliminar filas con valores nulos en las columnas de notas
    df.dropna(inplace=True)  # Los 'none' como str no se eliminarian

    # 6. Eliminar registros con notas anomalas (menores a 0 o mayores a 100)
    for col in ["math_score", "reading_score", "writing_score"]:
        df = df[(df[col] >= 0) & (df[col] <= 100)]

    return df
````

### 1.3.1 Transformación de notas a "Nota Chilena" 🇨🇱 

En una de las reuniones con Don Caguayo, este les comenta su intención de pasar todas sus notas a la "Escala Chilena". Para esto les facilita la siguiente ecuación:

$$Nota \, Chilena= \% \, de \, logro \times 0.06 + 1$$

Utilizando la ecuación entregada, cree una función que le permita transformar las notas (que están en escala 0-100) a la escala chilena [1,7] y entregue las notas redondeadas al segundo decimal.

**To-Do**

- [ ] Implementa la función `to_chilean_grade` que toma un **DataFrame** y el nombre de una **columna** con notas en escala 0-100, y retorna una **Serie** con las notas transformadas a escala chilena (rango 1.0-7.0). [0.3 puntos]
- [ ] Agrega las nuevas series de notas de matemáticas (columna `math_score_cl`), lectura (columna `reading_score_cl`) y escritura (columna `writing_score_cl`) al dataframe `df_cleaned`. [0.3 puntos]


**Respuesta Código:**

````python
# En esta celda implementa la operación solicitada
def to_chilean_grade(df: pd.DataFrame, columna: str) -> pd.Series:
    """Se deben crear tres nuevas columnas, la funcion to_chilean_grade recibe el dataframe y
    el nombre de una columna y retorna una serie de pandas con esa serie de pandas con esa columna transformada"""

    nota_chilena = df[columna].apply(lambda x: x * 0.06 + 1)
    return nota_chilena
````

````python
# En esta celda ejecuta la función y luego usa indexadores para agregar las nuevas columnas al dataframe limpio
math_chilean = to_chilean_grade(df_cleaned, "math_score")
writing_chilean = to_chilean_grade(df_cleaned, "writing_score")
reading_chilean = to_chilean_grade(df_cleaned, "reading_score")

# Indexador para agregar las nuevas columnas al dataframe limpio
df_cleaned["math_score_cl"] = math_chilean
df_cleaned["writing_score_cl"] = writing_chilean
df_cleaned["reading_score_cl"] = reading_chilean
````

### 1.3.2 Promedios 📋 [0.3 puntos]

Con todas las notas transformadas, la prestigiosa Universidad De la Cachaña necesita obtener el promedio general (GPA) de cada uno de los alumnos. Para esto se le solicita que calcule el promedio de las tres asignaturas y lo anexe como una nueva columna al DataFrame. 

Cabe señalar que las notas promedio deben ser **redondeadas** al segundo decimal.

**To-Do:**
- [ ] Implementa la función `calculate_average` que dado un dataframe con las notas por alumno, calcula y retorna dos **Series**: una con el promedio de las notas originales (escala 0-100) y otra con el promedio de las notas chilenas.
- [ ] Ejecuta la función y luego, usando indexadores apropiados, agrega las columnas `mean` (promedio escala 0-100) y `mean_cl` (promedio escala chilena) a `df_cleaned`.


**Respuesta:**

````python
# En esta celda implementa la operación solicitada
def calculate_average(df: pd.DataFrame) -> tuple[pd.Series, pd.Series]:
    """Comprobar que existen las columnas math_score_cl, writing_score_cl y reading_score_cl,
    luego calcular el promedio de estas tres columnas para cada fila y retornar una serie de pandas con el resultado
    Retorna dos series:
        - Una con el promedio de las notas originales
        - Otra con el promedio de las notas chilenas
    """
    if not all(col in df.columns for col in ["math_score_cl", "writing_score_cl", "reading_score_cl"]):
        raise ValueError(
            "Las columnas math_score_cl, writing_score_cl y reading_score_cl deben existir en el dataframe"
        )

    promedio_original = df[["math_score", "writing_score", "reading_score"]].mean(axis=1)
    promedio_chileno = df[["math_score_cl", "writing_score_cl", "reading_score_cl"]].mean(axis=1)

    return promedio_original, promedio_chileno
````

````python
# En esta celda ejecuta la función y luego usa indexadores para agregar la nueva columna al dataframe limpio
promedio_original, promedio_chileno = calculate_average(df_cleaned)

# Indexador para agregar la nueva columna al dataframe limpio
df_cleaned["mean"] = promedio_original
df_cleaned["mean_cl"] = promedio_chileno
````

### 1.3.3 Rankings 📶 [0.4 puntos]

¡Excelente trabajo calculando los promedios! Ahora Don Caguayo quiere saber quiénes son los estudiantes top y generar un ranking general.

**To-Do:**
- [ ] Implementa la función `rank_students` que dado un dataframe con las notas por alumno, obtiene una **Serie** con el ranking de alumnos, de mejor a peor nota (usando el promedio chileno `mean_cl`).
- [ ] Ejecuta la función y luego, usando indexadores apropiados, agrega la columna `rank` al `df_cleaned`.


````python
# En esta celda implementa la operación solicitada
def rank_students(df: pd.DataFrame) -> pd.Series:
    """
    Genera el ranking como serie de pandas para indexarlo luego al dataframe limpio, el ranking se debe generar a partir de la columna mean_cl
    """
    if "mean_cl" not in df.columns:
        raise ValueError("La columna mean_cl debe existir en el dataframe")

    # Generar ranking
    ranking = df["mean_cl"].rank(ascending=False, method="min").astype(int)

    return ranking
````

### 1.3.5 Padres Helicóptero 👨‍👩‍👧

Algunos padres (extremadamente entrometidos) preocupados por el desempeño de sus hijos, desean saber si estos han aprobado o no el año. Como la tarea de buscar a mano cada uno de estos alumnos en el `DataFrame` resulta demasiado tedioso, el rector le solicita crear una función de consulta. 

Cree una función que:

1. Reciba el `DataFrame` y una lista con nombres de alumnos
2. Extraiga las notas chilenas de matemáticas, escritura, lectura y promedio general (`mean_cl`) de los alumnos listados en `student_list` [0.3 puntos]
3. Genere una nueva columna transitoria `is_failing` (**asegurarse de no agregarla al dataframe original**) cuyo valor sea `True` si el alumno está reprobando o `False` en caso contrario [0.3 puntos]
4. Ordene por promedio chileno, de mayor a menor [0.2 puntos]
5. Reinicie el índice para que este comience desde 0 [0.2 puntos]


**Hint:** Puede serle de utilidad utilizar `isin()`. Recuerde además usar una copia del dataframe original para no modificarlo con la nueva columna.

**Respuesta:**

````python
def query_students(df: pd.DataFrame, student_list: list[str]) -> pd.DataFrame:
    # 1. Se mantienen las columnas match_score_cl, writing_score_cl, reading_score_cl, mean_cl y rank
    df_students = df.copy()
    df_students = df[df["names"].isin(student_list)][
        ["names", "math_score_cl", "writing_score_cl", "reading_score_cl", "mean_cl"]
    ]

    # 2. Columna transitoria is_failing (True si mean_cl < 4.0, False en caso contrario)
    df_students["is_failing"] = df_students["mean_cl"] < 4.0

    # 3. Ordenar el dataframe por la columna mean_cl de mayor a menor
    df_students = df_students.sort_values("mean_cl", ascending=False)

    # 4. Reiniciar indices para que vayan de 0 a n-1
    df_students.reset_index(drop=True, inplace=True)

    return df_students
````

### 1.3.6 Categorías de Rendimiento 📊

El rector, intrigado por saber si existe una relación entre el nivel educacional de los padres y las notas de sus hijos, les plantea estudiar dicha relación mediante categorización de los datos. 

Para esto, se les solicita que implementen la función `get_categories`, donde se generen dos series con diferentes tipos de categorización a partir de los promedios chilenos (`mean_cl`):

- Una con `pd.cut` con las labels `["B1", "B2", "B3", "B4", "B5"]` [0.2 Puntos]
- Otra con `pd.qcut`, con las labels `["Q1", "Q2", "Q3", "Q4", "Q5"]` [0.2 Puntos]

La función debe retornar una tupla con ambas series. Después, usando indexadores apropiados, agreguen a `df_cleaned` las columnas `cut_cl` y `qcut_cl`.

**Respuesta:**

````python
# En esta celda implementa la operación solicitada
def get_categories(df: pd.DataFrame) -> tuple[pd.Series, pd.Series]:
    """Se generan 2 series con diferentes tipos de categorización a partir de los promedios 'mean_cl'ArithmeticError
    - Una con pd.cut con labels ['B1', 'B2', 'B3', 'B4', 'B5']
    - Otra con pd.qcut con labels ['Q1', 'Q2', 'Q3', 'Q4', 'Q5']
    """
    # pd.cut con labels ['B1', 'B2', 'B3', 'B4', 'B5']
    bins = [1, 2.2, 3.4, 4.6, 5.8, 7.0]
    labels_cut = ["B1", "B2", "B3", "B4", "B5"]
    categories_cut = pd.cut(df["mean_cl"], bins=bins, labels=labels_cut)

    # pd.qcut con labels ['Q1', 'Q2', 'Q3', 'Q4', 'Q5']
    labels_qcut = ["Q1", "Q2", "Q3", "Q4", "Q5"]
    categories_qcut = pd.qcut(df["mean_cl"], q=5, labels=labels_qcut)

    return categories_cut, categories_qcut
````

````python
categorias_cut, categorias_qcut = get_categories(df_cleaned)
df_cleaned["cut_cl"] = categorias_cut
df_cleaned["qcut_cl"] = categorias_qcut
````

> - ¿Qué diferencias hay entre las categorías generadas por `cut` y `qcut` (en términos conceptuales rangos-percentiles)?


**Respuesta 📝:**

```
Las categorías generadas por qcut representan cuantiles, en los cuales cada grupo tiene aproximadamente el mismo número de estudiantes, esto a su vez genera que los rangos de las notas para cada cuantil varien entre sí.

Por otro lado, cut simplemente divide por rangos fijos o umbrales definidos, sin considerar la distribución de las notas.
```
